# Linear Regression (10000 iterations & 5 parameters)

## Data Cross validation

```python
## convert our data to 'list' dtype
xies = list(data['x'])
yies = list(data['y'])
## Data split coefficient ==>   Train_set: 0.7  ,   Test_set: 0.2   ,   Cross_validation: 0.1
## amount od our dataset : 300 rows
## split xies
Train_set_x_index , Test_set_x_index , Cross_validation_x_index = 210, 60 , 30
Train_set_x , Test_set_x , Cross_validation_x = list() , list() , list()

for i in range(0 , 210 , 1):
    Train_set_x.append(xies[i])

for i in range(210 , 270 , 1):
    Test_set_x.append(xies[i])

for i in range(270 , 300 , 1):
    Cross_validation_x.append(xies[i])

## split yies
Train_set_y_index , Test_set_y_index , Cross_validation_y_index = 210, 60 , 30
Train_set_y , Test_set_y , Cross_validation_y = list() , list() , list()
for i in range(0 , 210 , 1):
    Train_set_y.append(yies[i])

for i in range(210 , 270 , 1):
    Test_set_y.append(yies[i])

for i in range(270 , 300 , 1):
    Cross_validation_y.append(yies[i])
```

```python
## Gradiant Descent with 5 parameter and 10000 iteration
result = GD.GD5(Train_set_x , Train_set_y)
```

## Train_Set LinearRegression

```python
a = result[0]
b = result[1]
c = result[2]
d = result[3]
e = result[4]
f = result[5]
my_xies = list()
my_yies = list()
step = 2 / 100
temp = -1
for i in range(1 , 100 , 1):
    temp = temp+step
    y = GD.LinearHypoThesis5(a , b , c , d , e , f , temp)
    my_xies.append(temp)
    my_yies.append(y)
print("Cost value:" + str(GD.costFunction5(Train_set_x , Train_set_y , a , b , c , d , e , f)))
plt.plot(Train_set_x , Train_set_y ,'o' , my_xies , my_yies )
plt.xlabel('Train_set_x')
plt.ylabel('Train_set_y')
plt.title('Trainig data set Linear regression')
plt.show()
```

```python
cost_list = list(result[6])
iter = list(result[7])
plt.plot(iter , cost_list)
plt.xlabel('Iteration')
plt.ylabel('Cost Function')
plt.title('Garadiant Descent Performance')
```

## Test_Set LinearRegression

```python
a = result[0]
b = result[1]
c = result[2]
d = result[3]
e = result[4]
f = result[5]
my_xies = list()
my_yies = list()
step = 2 / 100
temp = -1
for i in range(1 , 100 , 1):
    temp = temp+step
    y = GD.LinearHypoThesis5(a , b , c , d , e , f , temp)
    my_xies.append(temp)
    my_yies.append(y)
print("Cost value:" + str(GD.costFunction5(Test_set_x , Test_set_y , a , b , c , d , e , f)))
plt.plot(Test_set_x , Test_set_y ,'o' , my_xies , my_yies )
plt.xlabel('Test_set_x')
plt.ylabel('Test_set_y')
plt.title('Testing data set Linear regression')
plt.show()
```
