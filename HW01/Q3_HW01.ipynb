##### A) Integer Encoding: Where each unique label is mapped to an integer.
##### One Hot Encoding: Where each label is mapped to a binary vector.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder

ds = pd.read_csv('test.csv')
```

```python
from sklearn import preprocessing
 
label_encoder = preprocessing.LabelEncoder()
 

ds['gender']= label_encoder.fit_transform(ds['gender'])
ds['gender'].unique()

ds['smoker']= label_encoder.fit_transform(ds['smoker'])
ds['smoker'].unique()
```

##### C) integer encoding is a natural encoding for ordinal variables but for categorical variables, it imposes an ordinal relationship where no such relationship may exist. This can cause problems and a one-hot encoding may be used instead.
