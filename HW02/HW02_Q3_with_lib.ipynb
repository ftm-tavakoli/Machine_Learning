```python
import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
# read in the iris data
iris = load_iris()

# create X (features) and y (response)
x = iris.data
y = iris.target
```

```python
# search for an optimal value of K for KNN

k_range = range(1, 31)
k_scores = []

#def KNN(distanceFunction, k, dataSet, instance)
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x, y, cv=10, scoring='accuracy')
    k_scores.append(scores.mean())
print('Length of list', len(k_scores))
print('Max of list', max(k_scores))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-validated accuracy')
```

```python
knn = KNeighborsClassifier(n_neighbors=20)
print(cross_val_score(knn, x, y, cv=10, scoring='accuracy').mean())
```

```python
knn = KNeighborsClassifier(n_neighbors=20)
print(cross_val_score(knn, x, y, cv=10, scoring='neg_mean_squared_error').mean())
print(cross_val_score(knn, x, y, cv=10, scoring='neg_root_mean_squared_error').mean())
print(cross_val_score(knn, x, y, cv=10, scoring='neg_mean_absolute_error').mean())
```
