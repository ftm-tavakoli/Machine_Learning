```python
from sklearn.datasets import load_iris
%matplotlib inline
import pylab
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = load_iris()
f = data['data']
f_name = data['feature_names']
target = data['target']
```

```python
pylab.rcParams['figure.figsize'] = (5.0, 8.0)


markers, colors  = "+.*", "ckr"
pairs = [(0,1),(2,3)]
for i,(x,y) in enumerate(pairs):
    plt.subplot(2,1,i+1)
    for cls,marker,clr in zip(range(3),markers,colors):
        plt.scatter(f[target == cls,x], f[target == cls,y], marker=marker, c=clr)
    p_f,p_cls  = f[-90],target[-90]
    marker,clr = markers[p_cls], colors[p_cls]
    plt.scatter(p_f[x],p_f[y],marker=marker,c=clr,s=100)
    plt.xlabel(f_name[x])
    plt.ylabel(f_name[y])
```
